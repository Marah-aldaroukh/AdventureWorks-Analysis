```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.boxplot (x=df['Profit'])
plt.title('Boxplot of Profit')
plt.show()
```

```python
Q1= df['Profit'].quantile(0.25)
Q3= df['Profit'].quantile(0.75)
IQR=Q3-Q1
lower_bound=Q1-1.5*IQR
upper_bound=Q3+1.5*IQR
outliers=df[(df['Profit']<lower_bound)|(df['Profit']>upper_bound)]
print (f"Number of outliers in Profit: {len(outliers)}")
print (f"The percentage:{len(outliers)/len(df)*100:.2f}%")
```

```python
financial_cols=['Revenue','OrderQuantity','ProductPrice','ProductCost']
fig,axes= plt.subplots(2,2,figsize=(12,8))
axes=axes.flatten()
for i,col in enumerate(financial_cols):
    sns.boxplot(x=df[col],ax=axes[i])
    axes[i].set_title(f'Boxplot of {col}')
plt.tight_layout()
plt.show()
```

```python
financial_outliers = []

for col in ['Revenue', 'ProductPrice', 'ProductCost']:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    outliers = df[(df[col] < lower) | (df[col] > upper)]
    pct = (len(outliers) / len(df)) * 100
    financial_outliers.append({
        'Column': col,
        'Outliers Count': len(outliers),
        'Percentage': f'{pct:.2f}%'
    })

outliers_df = pd.DataFrame(financial_outliers)
print(outliers_df)
```

```python
Q1 = df['ProductPrice'].quantile(0.25)
Q3 = df['ProductPrice'].quantile(0.75)
IQR = Q3 - Q1
upper_bound = Q3 + 1.5 * IQR

print(f"The maximum price of the product: {upper_bound}")

normal_products = df[df['ProductPrice'] <= upper_bound]
premium_products = df[df['ProductPrice'] > upper_bound]

print(f"the number of normal products: {len(normal_products)}")
print(f"the number of premium products: {len(premium_products)}")
print(f"the percentage of premium: {len(premium_products)/len(df)*100:.2f}%")
```

```python
print("normal products statistics:")
print(normal_products[['Profit', 'Revenue', 'ProductPrice']].describe())

print("\npremium products statistics:")
print(premium_products[['Profit', 'Revenue', 'ProductPrice']].describe())
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.hist(normal_products['Profit'], bins=50, color='blue', alpha=0.7)
plt.title('Distribution of normal product profits')

plt.subplot(1,2,2)
plt.hist(premium_products['Profit'], bins=50, color='green', alpha=0.7)
plt.title('Distribution of premium product profits')
plt.tight_layout()
plt.show()
```

```python
total_profit_normal = normal_products['Profit'].sum()
total_profit_premium = premium_products['Profit'].sum()
total_profit_all = df['Profit'].sum()

print(f"Total profit (normal products): {total_profit_normal:,.2f}")
print(f"Total profit (premium products): {total_profit_premium:,.2f}")
print(f"Total profit (all products): {total_profit_all:,.2f}")

print(f"\nNormal products contribution: {total_profit_normal/total_profit_all*100:.2f}%")
print(f"Premium products contribution: {total_profit_premium/total_profit_all*100:.2f}%")
```

```python
margin_normal = (normal_products['Profit'].sum() / normal_products['Revenue'].sum()) * 100
margin_premium = (premium_products['Profit'].sum() / premium_products['Revenue'].sum()) * 100

print(f"Profit Margin (Normal Products): {margin_normal:.2f}%")
print(f"Profit Margin (Premium Products): {margin_premium:.2f}%")
```

```python
top_profit_products = df.groupby('ProductName')['Profit'].sum().sort_values(ascending=False).head(10)
print("top profit products:")
print(top_profit_products)

import matplotlib.pyplot as plt
plt.figure(figsize=(10,5))
top_profit_products.plot(kind='bar', color='green')
plt.title('Top 10 Most Profitable Products')
plt.ylabel('Total Profit')
plt.xlabel('Product Name')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Top 10 best‑selling products by quantity (all years)
top_selling_all = df.groupby('ProductName')['OrderQuantity'].sum().sort_values(ascending=False).head(10)

print("Top 10 best‑selling products (all years):")
print(top_selling_all)

# Optional: simple bar chart
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
top_selling_all.plot(kind='bar', color='skyblue')
plt.title('Top 10 Best‑Selling Products (All Years)')
plt.ylabel('Total Quantity Sold')
plt.xlabel('Product Name')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
df['Year'] = df['OrderDate'].dt.year
df['Month'] = df['OrderDate'].dt.month

monthly_summary = df.groupby(['Year', 'Month']).agg(
    Total_Revenue=('Revenue', 'sum'),
    Total_Profit=('Profit', 'sum'),
    Order_Count=('OrderNumber', 'nunique')  # عدد الطلبات الفريدة
).reset_index()

monthly_summary['YearMonth'] = monthly_summary['Year'].astype(str) + '-' + monthly_summary['Month'].astype(str).str.zfill(2)

print(monthly_summary)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(14,6))
plt.plot(monthly_summary['YearMonth'], monthly_summary['Total_Revenue'], marker='o', label='Revenue', linewidth=2)
plt.plot(monthly_summary['YearMonth'], monthly_summary['Total_Profit'], marker='s', label='Profit', linewidth=2)
plt.title('Monthly Revenue and Profit (All Products)')
plt.xlabel('Year-Month')
plt.ylabel('Amount')
plt.xticks(rotation=45, ha='right')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
yearly_actual = df.groupby('Year').agg(
    Total_Revenue=('Revenue', 'sum'),
    Total_Profit=('Profit', 'sum'),
    Order_Count=('OrderNumber', 'nunique')
).reset_index()

print(yearly_actual)
```

```python
# December analysis
december_data = df[df['Month'] == 12]

# Top 10 products by profit in December
top_profit_dec = december_data.groupby('ProductName')['Profit'].sum().sort_values(ascending=False).head(10)

# Top 10 products by quantity sold in December
top_quantity_dec = december_data.groupby('ProductName')['OrderQuantity'].sum().sort_values(ascending=False).head(10)

print("Top 10 most profitable products in December:")
print(top_profit_dec)

print("\nTop 10 best-selling products (by quantity) in December:")
print(top_quantity_dec)

# Check overlap between the two lists
common_products = set(top_profit_dec.index) & set(top_quantity_dec.index)
print(f"\nProducts appearing in both lists: {len(common_products)}")
print(common_products)
```

```python
# Segment analysis in December
dec_data = december_data.copy()
dec_data['Segment'] = dec_data['ProductPrice'].apply(lambda x: 'Premium' if x > upper_bound else 'Normal')

# Revenue and profit by segment
segment_summary = dec_data.groupby('Segment').agg(
    Total_Revenue=('Revenue', 'sum'),
    Total_Profit=('Profit', 'sum')
).round(2)

print("\nSegment contribution in December:")
print(segment_summary)
```

```python

dec_segment_summary = dec_data.groupby('Segment').agg(
    Total_Quantity=('OrderQuantity', 'sum'),
    Total_Revenue=('Revenue', 'sum'),
    Total_Profit=('Profit', 'sum')
).round(2)

dec_segment_summary['Profit_Share_%'] = (dec_segment_summary['Total_Profit'] / dec_segment_summary['Total_Profit'].sum()) * 100
dec_segment_summary['Quantity_Share_%'] = (dec_segment_summary['Total_Quantity'] / dec_segment_summary['Total_Quantity'].sum()) * 100

print(dec_segment_summary)
```

```python
returns_data = df[df['ReturnQuantity'].notna()]
print(f"Number of returns operations: {len(returns_data)}")
print(f"Number of returned products: {returns_data['ProductName'].nunique()}")
```

```python
top_returned_products = returns_data.groupby('ProductName')['ReturnQuantity'].sum().sort_values(ascending=False).head(10)

print("Top 10 products returned by quantity):")
print(top_returned_products)

import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
top_returned_products.plot(kind='bar', color='red')
plt.title('Top 10 Returned Products (by Quantity)')
plt.ylabel('Total Return Quantity')
plt.xlabel('Product Name')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
returns_data['ReturnLoss'] = returns_data['ReturnQuantity'] * returns_data['ProductPrice']
top_loss_products = returns_data.groupby('ProductName')['ReturnLoss'].sum().sort_values(ascending=False).head(10)

print("Top 10 products with the most returns losses:")
print(top_loss_products)

plt.figure(figsize=(10,5))
top_loss_products.plot(kind='bar', color='darkred')
plt.title('Top 10 Products by Return Loss')
plt.ylabel('Total Return Loss')
plt.xlabel('Product Name')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
product_sales = df.groupby('ProductName').agg(
    Total_Sold=('OrderQuantity', 'sum'),
    Total_Returned=('ReturnQuantity', 'sum')
).reset_index()

product_sales['Return_Rate_%'] = (product_sales['Total_Returned'] / product_sales['Total_Sold']) * 100

top_return_rate = product_sales.sort_values('Return_Rate_%', ascending=False).head(10)

print("Top 10 products in return rate:")
print(top_return_rate[['ProductName', 'Total_Sold', 'Total_Returned', 'Return_Rate_%']])
```

```python
product_performance = df.groupby('ProductName').agg(
    Total_Profit=('Profit', 'sum'),
    Total_Returned=('ReturnQuantity', 'sum'),
    Total_Sold=('OrderQuantity', 'sum')
).reset_index()

product_performance['Return_Rate_%'] = (product_performance['Total_Returned'] / product_performance['Total_Sold']) * 100

top_performers = product_performance.sort_values('Total_Profit', ascending=False).head(20)

print("Top 20 products with return rate:")
print(top_performers[['ProductName', 'Total_Profit', 'Return_Rate_%']].round(2))
```
